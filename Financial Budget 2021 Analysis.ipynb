```python
#Import libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Import data

df=pd.read_csv('C:/Users/91782/Documents/Python Scripts/India_budget_2021.csv')
df.head()
```

```python
#Check the number of rows and columns

df.shape
```

```python
#Check null values

df.isna().sum()
```

```python
#Delete null values

data=df.dropna()
```

```python
#Verify null values are removed
data.isna().sum()
```

```python
#Bar Graph

data.plot.bar(x='Department /Ministry',y='Fund allotted(in ₹crores)',figsize=(15,10))
```

```python
#Remove Grand Total

data=data.iloc[:-1,:]
```

```python
#Verify Grand total is deleted

data.plot.bar(x='Department /Ministry',y='Fund allotted(in ₹crores)',figsize=(15,10))
```

```python
#Sort from largest to smallest
data_sorted=data.sort_values(by='Fund allotted(in ₹crores)',ascending=False)
data_sorted
```

```python
#Plot Top 10 Budget Ministries

data_sorted.head(10).plot.bar(x='Department /Ministry',y='Fund allotted(in ₹crores)',figsize=(15,10))

plt.title('Distribution of Top 10 budget ministries')

plt.show()
```

```python
#Plot Last 10 Budget Ministries

data_sorted.tail(10).plot.bar(x='Department /Ministry',y='Fund allotted(in ₹crores)',figsize=(15,10))

plt.title('Distribution of Last 10 budget ministries')

plt.show()
```

```python
#Pie Chart - Top 10 Budget Ministries

new_data=data_sorted.head(10)

df=new_data['Fund allotted(in ₹crores)']
labels=new_data['Department /Ministry']

plt.figure(figsize=(10,9))

plt.pie(df,labels=labels,autopct='%1.1f%%',startangle=45,pctdistance=0.7,shadow=True)
central_circle=plt.Circle((0,0),0.5,color='white')

fig=plt.gcf()

fig.gca().add_artist(central_circle)
plt.rc('font',size=12)

plt.title('Distribution of Top 10 Budget Ministries',fontweight='bold')

plt.show()
```
